```python
X = []
y = []

images_dir = 'images'
labels_dir = 'manual1'

images = sorted(os.listdir(images_dir))
labels = sorted(os.listdir(labels_dir))

for img_dir, label_dir in zip(images, labels):
    img = cv2.imread(os.path.join(images_dir, img_dir), cv2.IMREAD_GRAYSCALE)
    label = cv2.imread(os.path.join(labels_dir, label_dir), cv2.IMREAD_GRAYSCALE)
    X.append(img)
    y.append(label)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,3))
ax[0].imshow(X[0], cmap='gray')
ax[1].imshow(y[0], cmap='gray')
plt.show()
```
